```python
import base64
from github import Github
from pprint import pprint
import calendar, time
import pandas as pd


# pygithub object
g = Github("ACCESS_API")

# get repositories
user = g.search_repositories(query="wikidata")
print(user.totalCount)
```

```python
def print_repo(repo, df, queryw):
    # add 5 seconds to be sure the rate limit has been reset
    #core_rate_limit = g.get_rate_limit().core
    #reset_timestamp = calendar.timegm(core_rate_limit.reset.timetuple())
    #sleep_time = reset_timestamp - calendar.timegm(time.gmtime()) - 3500 + 5

    sleep_time = 5
    print("Timeout: ", sleep_time, " seconds")
    time.sleep(sleep_time)
    # repository full name
    fullname = str(repo.full_name)
    # repository description
    desc = str(repo.description)
    # the date of when the repo was created
    date = str(repo.created_at)
    # the date of the last git push
    lastpush = str(repo.pushed_at)
    # home website (if available)
    homepage = str(repo.homepage)
    # programming language
    language = str(repo.language)
    # number of forks
    forks = str(repo.forks)
    # number of stars
    stars = str(repo.stargazers_count)
    try:
        # repo license
        license = base64.b64decode(repo.get_license().content.encode()).decode()
    except:
        license = ""
    license = license[:license.find("\n")].strip()
    # repository content (files & directories)
    filel = []
    for content in repo.get_contents(""):
        filel.append(content.path)
    files = "|".join(filel)
    # topics
    topicl = repo.get_topics()
    topics = "|".join(topicl)
    # issues
    issuel = []
    for issue in repo.get_issues():
        issuel.append(issue.title)
    issues = "|".join(issuel)
    print("-"*50)
    df.loc[len(df.index)] = [fullname, desc, date, lastpush, homepage, language, forks, stars, license, files, topics, issues, queryw]
    
```

```python
df = pd.DataFrame(columns=['Full name', 'Description', 'Date created', 'Date of last push', 'Home Page', 'Language', 'Number of forks', 'Number of stars', 'License', 'Files', 'Topics', 'Issues', 'Category'])
for repo in user:
    print_repo(repo, df, "Wikidata")
    print(repo)
```

#Collecting the metadata of the 1000 most relevant Wikipedia-related GitHub repositories

```python
# get repositories
user = g.search_repositories(query="wikipedia")
print(user.totalCount)


for repo in user:
    print_repo(repo, df, "Wikipedia")
    print(repo)
```
